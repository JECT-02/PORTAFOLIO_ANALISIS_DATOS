# 01 Data Exploration

Exploracion de datos bronze y visualizacion inicial.

```
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

plt.style.use('dark_background')
bronze_dir = Path('../data/bronze')
```

```
info_files = sorted(bronze_dir.glob('stock_info_*.json'))
with open(info_files[-1], 'r') as f:
    info = json.load(f)
info_df = pd.DataFrame(info)
print('Shape:', info_df.shape)
info_df.head()
```

```
info_df['sector'].value_counts().plot(kind='bar', figsize=(10,5), title='Distribucion por Sector')
plt.show()
```

```
hist_files = list((bronze_dir / 'historical').glob('*.json'))
print(f'Total historical files: {len(hist_files)}')
```
